```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming df is your pandas DataFrame containing the dataset

# Numeric Columns
numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns

# Categorical Columns
categorical_columns = df.select_dtypes(include=['object']).columns

# Univariate Analysis - Numeric Columns (Histograms)
for col in numeric_columns:
    plt.figure(figsize=(8, 5))
    sns.histplot(df[col], kde=True)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()

# Univariate Analysis - Categorical Columns (Countplots)
for col in categorical_columns:
    plt.figure(figsize=(8, 5))
    sns.countplot(data=df, x=col)
    plt.title(f'Count of categories in {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.show()
```

```python
# Bivariate Analysis - Numeric vs Numeric (Scatter Plot and Correlation)
plt.figure(figsize=(10, 8))
sns.scatterplot(data=df, x='alcohol', y='quality')
plt.title('Scatter plot of alcohol content vs quality')
plt.xlabel('Alcohol content')
plt.ylabel('Quality')
plt.show()

# Correlation Matrix for Numerical Variables
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()

# Bivariate Analysis - Categorical vs Numeric (Box Plot)
plt.figure(figsize=(10, 8))
sns.boxplot(data=df, x='quality', y='alcohol')
plt.title('Box plot of alcohol content across quality levels')
plt.xlabel('Quality')
plt.ylabel('Alcohol content')
plt.show()

# Bivariate Analysis - Categorical vs Categorical (Heatmap)
plt.figure(figsize=(10, 8))
pivot = df.groupby(['quality']).size().reset_index().pivot(columns='quality', values=0)
sns.heatmap(pivot, annot=True, cmap='Blues', linewidths=0.5)
plt.title('Heatmap of wine quality by color')
plt.xlabel('Quality')
plt.ylabel('Color')
plt.show()
```

```python
from  sklearn.preprocessing import StandardScaler
```

```python
x =df.drop(["quality"],axis=1,inplace =False)
y = df.quality
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size =.30,random_state=42)
print(x_train.shape)
print(x_test.shape)
print(y_test.shape)
print(y_train.shape)
```

```python
from sklearn.linear_model import ElasticNet
lr =ElasticNet(alpha= 0.1,l1_ratio=0.5,max_iter=1129,random_state=42)
lr.fit(x_train,y_train)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, mean_squared_error, r2_score,mean_absolute_error
import numpy as np
```

```python
def eval_metrics(actual,predicted):
    rmse = np.sqrt(mean_squared_error(actual,predicted))
    mae = mean_absolute_error(actual,predicted)
    r2 = r2_score(actual,predicted)

    return rmse,mae,r2
```

```python
predict_value = lr.predict(x_test)
predict_value
```

```python
(rmse,mae,r2) = eval_metrics(y_test,predict_value)
print("rmae \t ", rmse)
print("mae \t",mae)
print("r2 \t ", r2)
```
